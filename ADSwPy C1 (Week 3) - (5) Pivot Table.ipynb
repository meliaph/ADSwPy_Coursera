## **(Week 3) - (5) Pivot Table**

```
#Importing the Libraries
import pandas as pd
import numpy as np
```

```
#Importing the Dataset
df=pd.read_csv('/content/cwurData.csv')
df.sample(n=7)
```

```
#Create a Function for Creating the Category
def create_category(ranking):

  #Category 1: First Tier Top Unversity
  if (ranking >= 1) & (ranking<=100):
    return "First Tier Top Unversity"

  #Category 2: Second Tier Top Unversity
  elif (ranking >= 101) & (ranking<=200):
    return "Second Tier Top Unversity"
  
  #Category 3: Third Tier Top Unversity
  elif (ranking >= 201) & (ranking<=300):
    return "Third Tier Top Unversity"
  
  #The rest
  return "Other Top Unversity"
```

```
#Create a Column that Indicate the Category: "Rank Level"
df['Rank_Level']=df['world_rank'].apply(lambda x: create_category(x))
```

```
df.sample(n=10)
```

```
#Pivoting the Table 
df.pivot_table(values='score',index='country',columns='Rank_Level',aggfunc=[np.mean]).head(n=10)
```

```
df.pivot_table(values='score',index='country',columns='Rank_Level',aggfunc=[np.mean, np.max]).head(n=10)
```

```
df.pivot_table(values='score',index='country',columns='Rank_Level',
               aggfunc=[np.mean, np.max],margins=True).head(10)
```

```
new_df=df.pivot_table(values='score',index='country',columns='Rank_Level',
                      aggfunc=[np.mean, np.max],margins=True)
```
