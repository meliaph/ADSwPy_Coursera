## **Merging Data Frame**

```
import pandas as pd
```

#### **SET INDEX BY NAME**

```
#Staff DF
staff_df=pd.DataFrame([{'Name':'Kelly','Role':'Director of HR'},
                       {'Name':'Sally','Role':'Course liasion'},
                       {'Name':'James','Role':'Grader'}])

#Student DF
student_df=pd.DataFrame([{'Name':'James','School':'Business'},
                         {'Name':'Mike','School':'Law'},
                         {'Name':'Sally','School':'Engineering'}])
```

```
#Set index by column 'Name'
staff_df=staff_df.set_index('Name')
student_df=student_df.set_index('Name')
```

```
staff_df
```

```
student_df
```

```
#Merging Outer
pd.merge(staff_df,student_df,how='outer',left_index=True,right_index=True)
```

```
#Merging Inner
pd.merge(staff_df,student_df,how='inner',left_index=True,right_index=True)
```

```
#Merging Left
pd.merge(staff_df,student_df,how='left',left_index=True,right_index=True)
```

```
#Merging Right
pd.merge(staff_df,student_df,how='right',left_index=True,right_index=True)
```

#### **SET INDEX AS NONE**

```
#Set index by column 'Name'
staff_df=staff_df.reset_index()
student_df=student_df.reset_index()
```

```
staff_df
```

```
student_df
```

```
#Merging Using 'on' Parameter
pd.merge(staff_df,student_df,how='outer',on='Name')
```

```
#Merging Inner
pd.merge(staff_df,student_df,how='inner',on='Name')
```

```
#Merging Left
pd.merge(staff_df,student_df,how='left',on='Name')
```

```
#Merging Right
pd.merge(staff_df,student_df,how='right',on='Name')
```

#### **CONFLICTS BETWEEN MERGED DATAFRAMES**

```
staff_df=pd.DataFrame([{'Name':'Kelly','Role':'Director of HR','Location':'State Street'},
                       {'Name':'Sally','Role':'Course liasion','Location':'Washington Avenue'},
                       {'Name':'James','Role':'Grader','Location':'Washington Avenue'}])
student_df=pd.DataFrame([{'Name':'James','School':'Business','Location':'1024 Billiard Avenue'},
                         {'Name':'Mike','School':'Law','Location':'Fraternity House #22'},
                         {'Name':'Sally','School':'Engineering','Location':'512 Wilson Crescent'}])
```

```
#Merging Using 'on' Parameter
pd.merge(staff_df,student_df,how='outer',on='Name')
```

```
#Merging Inner
pd.merge(staff_df,student_df,how='inner',on='Name')
```

```
#Merging Left
pd.merge(staff_df,student_df,how='left',on='Name')
```

```
#Merging Right
pd.merge(staff_df,student_df,how='right',on='Name')
```

##### **With 2 Parameters**

```
staff_df=pd.DataFrame([{'First Name':'Kelly','Last Name':'Desjardins','Role':'Director of HR'},
                       {'First Name':'Sally','Last Name':'Brooks','Role':'Course liasion'},
                       {'First Name':'James','Last Name':'Wilde','Role':'Grader'}])
student_df=pd.DataFrame([{'First Name':'James','Last Name':'Hammond','School':'Business'},
                         {'First Name':'Mike','Last Name':'Smith','School':'Law'},
                         {'First Name':'Sally','Last Name':'Brooks','School':'Engineering'}])
```

```
#Merging Using 'on' Parameter
pd.merge(staff_df,student_df,how='outer',on=['First Name','Last Name'])
```

```
#Merging Inner
pd.merge(staff_df,student_df,how='inner',on=['First Name','Last Name'])
```

```
#Merging Left
pd.merge(staff_df,student_df,how='left',on=['First Name','Last Name'])
```

```
#Merging Right
pd.merge(staff_df,student_df,how='right',on=['First Name','Last Name'])
```
