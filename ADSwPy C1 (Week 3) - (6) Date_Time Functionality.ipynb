####**DatetimeIndex and PeriodIndex**

```
t1=pd.Series(list('abc'),[pd.Timestamp('2016-09-01'),
                          pd.Timestamp('2016-09-02'),
                          pd.Timestamp('2016-09-03')])

t1
```

```
type(t1.index)
```

```
t2=pd.Series(list('def'),[pd.Period('2016-09'),
                          pd.Period('2016-10'),
                          pd.Period('2016-11')])
t2
```

```
type(t2.index)
```

####**Converting to Datetime**

```
d1=['2 June 2013','Aug 29, 2014','2015-06-26','7/12/16']
d1
```

```
ts3=pd.DataFrame(np.random.randint(10,100,(4,2)),index=d1,columns=list('ab'))
ts3
```

```
ts3.index=pd.to_datetime(ts3.index)
ts3
```

```
pd.to_datetime('4.7.12',dayfirst=True)
```

####**Timedelta**

```
pd.Timestamp('12/31/2016')-pd.Timestamp('9/16/2016')
```

```
pd.Timestamp('9/2/2016 8:10AM')+pd.Timedelta('12D 3H')
```

####**Offset**

```
pd.Timestamp('12/31/2020').weekday()

#It is Thursday = 3+1
#with .isoweekday() it will show 4 instead of 3
#pd.Timestamp('12/31/2020').isoweekday()
```

```
pd.Timestamp('9/4/2016')+pd.offsets.Week(2)
```

```
pd.Timestamp('9/4/2016')+pd.offsets.MonthEnd()
```

####**Working with Dates in a Dataframe**

```
dates=pd.date_range('10-01-2016',periods=9,freq='2W-SUN')
dates
```

```
pd.date_range('10-01-2016',periods=9,freq='B')
```

```
pd.date_range('04-01-2016',periods=12,freq='QS-JUN')
```

```
dates=pd.date_range('10-01-2016',periods=9,freq='2W-SUN')
df=pd.DataFrame({'Count 1':100+np.random.randint(-5,10,9).cumsum(),
                 'Count 2':120+np.random.randint(-5,10,9)},index=dates)
df
```

```
df.index.day_name()
```

```
df.diff()
```

```
df.resample('M').mean()
```
