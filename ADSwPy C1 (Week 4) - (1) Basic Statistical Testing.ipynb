## **(Week 4) - (1) Basic Statistical Testing**

**(Notes)** Hypothesis testing is a core data analysis activity behind the experimentation which goal is to determine if, for instance, the two different conditions we have in an experiment have resulted in different impacts.

Hypothesis testing consists of two statements of interest:
* Alternative hypothesis --> a statement that describes that there is a relationship between two selected variables in a study
* Null hypothesis --> general statement that states that there is no relationship between two phenomenons under consideration or that there is no association between two groups

These two statements are the two mutually exclusive statements about a population in hypothesis testing.

#####**Split that Assignment Submissions Data into 2 DataFrame**

```
#1st DataFrame: Assignment1 Finished Earlier than 2016
df_early=df[pd.to_datetime(df['assignment1_submission'])<'2016']
df_early.sample(n=3)
```

```
#Size of df_early
df_early.shape
```

```
#2nd DataFrame: Assignment1 Finished in 2016 Forward
df_later=df[~df.index.isin(df_early.index)]
df_later.sample(n=3)
```

```
#Size of df_later
df_later.shape
```

```
#The Average of Assignment 1 Grade for Early Submission and Later Submission
print("The Average of Assignment 1 Grade for Early Submission",df_early['assignment1_grade'].mean())
print("The Average of Assignment 1 Grade for Later Submission",df_later['assignment1_grade'].mean())
```

```
#Performing t-test using SciPy
from scipy.stats import ttest_ind
```

```
#Performing t-test to the Early Submission and Later Submission DataFrame
print("t-Test Results for Assignment 1:",ttest_ind(df_early['assignment1_grade'],df_later['assignment1_grade']))
print("t-Test Results for Assignment 2:",ttest_ind(df_early['assignment2_grade'],df_later['assignment2_grade']))
print("t-Test Results for Assignment 3:",ttest_ind(df_early['assignment3_grade'],df_later['assignment3_grade']))
print("t-Test Results for Assignment 4:",ttest_ind(df_early['assignment4_grade'],df_later['assignment4_grade']))
print("t-Test Results for Assignment 5:",ttest_ind(df_early['assignment5_grade'],df_later['assignment5_grade']))
print("t-Test Results for Assignment 6:",ttest_ind(df_early['assignment6_grade'],df_later['assignment6_grade']))
```

When the `pvalue` is higher than our alpha (0.05, because our confidence level is 95%), then we **CANNOT REJECT** the **Null Hypothesis**. That means, between Early Submission and Later Submission datasets, there is no relationship, or in other words, they **statistically insignificant**.
