**Q2**

We want to use the apply method to get a new DataFrame named result_df with a new column AVG. The AVG column should average the float values across P2010 to P2015. The apply method should also remove the 6 original columns (P2010 to P2015). For that, what should be the value of x and y in the given code?

```
x=1
y=1


frames=['P2010','P2011','P2012','P2013','P2014','P2015']
dfx=pd.DataFrame
dfx['AVG']=dfx[frames].apply(lambda z: np.mean(z),axis=x)
result_dfx=dfx.drop(frames,axis=y)
```
